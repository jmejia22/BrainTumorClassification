```python
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
import os
```

```python
def load_images(folder):
    """
    Loads in images from a folder as a list 
    """
    images = []
    for filename in os.listdir(folder):
        img = mpimg.imread(os.path.join(folder, filename))
        if img is not None:
            images.append(img)
    return images
```

```python
path = "/Users/wendytran/Downloads/Brain-Tumor-Classification-DataSet-master/Training/"
```

### Loading in Images

```python
glioma_tumor = load_images(path + "glioma_tumor")
len(glioma_tumor)
```

```python
meningioma_tumor = load_images(path + "meningioma_tumor")
len(meningioma_tumor)
```

```python
no_tumor = load_images(path + "no_tumor")
len(no_tumor)
```

```python
pituitary_tumor = load_images(path + "pituitary_tumor")
len(pituitary_tumor)
```
